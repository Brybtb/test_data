# View of Dividends in Accounts

```python
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
%matplotlib notebook
```

```python
# Read tax_test.csv
account_info = 'tax_test.csv'
tax_test = pd.read_csv(account_info, encoding = 'utf-8')
```

```python
# Strip whitespace/regex
tax_test.columns = tax_test.columns.str.replace("\r", "")

tax_test.columns = tax_test.columns.str.replace("\n", "")
```

```python
# Strip out commas
tax_test['Dollar DividendAmount'] = tax_test['Dollar DividendAmount'].str.replace(',', '')
```

```python
# Change Dollar Dividend Amount to numeric
tax_test['Dollar DividendAmount'] = pd.to_numeric(tax_test['Dollar DividendAmount'])
```

```python
# Create a new series using 'AccountType'
accounts = tax_test.groupby(['AccountType'])
```

```python
# Create a new series with dividend totals
div_totals = accounts['Dollar DividendAmount'].sum()
```

```python
# Create DataFrame off of groupby
div_per_account = pd.DataFrame(accounts['Dollar DividendAmount'].sum())
```

```python
# Reset index to make 'AccountType a column'
div_per_account = div_per_account.reset_index()
div_per_account.head()
```

```python
# Chart our data, give it a title, and label the axes
div_chart = div_totals.plot(kind="bar", title="Divs")
div_chart.set_xlabel("Account Type")
div_chart.set_ylabel("Dollars")

plt.show()
plt.tight_layout()
```

```python
# Set new index to 'AccountType'
div_sum = div_per_account.set_index("AccountType")
div_sum
```
